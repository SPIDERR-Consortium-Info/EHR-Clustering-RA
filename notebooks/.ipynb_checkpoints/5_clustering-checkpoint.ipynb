## #5.2 Create meta data for MAUI


```python
import pandas as pd
import numpy as np

l_columns = ['pseudoId', 'patnr', 'RF', 'aCCP', 'Sex', 'Age', 'Physician', 'deltaMan', 'MannTillTreat', 'DaysTillTreat', 'DAS28(3)', 'DAS44', 'firstMannequin', 'FirstConsult', 'symptomDuration', 'potentialRN']
l_cat = ['RF', 'aCCP', 'Sex']

# Get final patient selection (+ characteristics)
df_treat_new = pd.read_csv(r'../filters/RA_patients_AllNP_Characteristics_symptomDur.csv', sep='|')
print('Before selecting relevant patients: ')
print('PseudoId:\t\t', len(df_treat_new), '\t\nPatient id:\t\t', len(df_treat_new['patnr'].unique()))
l_pat_final = list(df_treat_new[df_treat_new['LEARNING_SELECTION']]['pseudoId'])
print('Final Patient selection:', len(l_pat_final))

df_treat_new = df_treat_new[l_columns]
df_treat_new['sample'] = [i for i in range(len(df_treat_new))]

# Get Raw lab values for metadata
df_lab = pd.read_csv(r'../new_data/3_wide/DF_Lab_wide_Final2.csv', sep='|')
new_lab_cols = {'MCV (fl)': "MCV", 'Hematocriet (L/L)' : "Ht",'Leukocyten (10^9/L)' : "Leuko", 'MCHC (mmol/L)' : "MCHC", 'MCH (fmol)' : "MCH", 'BSE (mm)' : "BSE", 'Hemoglobine (mmol/L)' : "Hb", 'Trombocyten (10^9/L)' : "Trom"} #  'Lymfocyten' : "Lym", 'Monocyten' : "Mono" 
df_lab = df_lab.rename(columns=new_lab_cols)
l_lab = list(new_lab_cols.values())
l_lab.append('pseudoId')
df_treat_new = df_treat_new.merge(df_lab[l_lab], on='pseudoId', how='left')

# Select relevant patients
df_treat_new = df_treat_new[df_treat_new['pseudoId'].isin(l_pat_final)]

df_meta = pd.read_csv('../new_data/4_processed/DF_Mannequin_NEW_Engineered.csv', sep='|')
l_keep = [x for x in list(df_meta.columns) if x not in ['patnr', 'FirstConsult']]
df_treat_new = df_treat_new.merge(df_meta[l_keep], on='pseudoId', how='left')
l_cat.extend(l_keep)

# Check if patients in EAC
df_eac = pd.read_csv(r'../new_data/offshoots/EAC_keys2.csv', sep=';')
d_eac = dict(zip(df_eac.member_entity_Pseudonym_value, df_eac.Identificatie_EACNUMM))
df_treat_new['IS_IN_EAC'] = df_treat_new['patnr'].apply(lambda x: 1 if x in d_eac.keys() and d_eac[x] == d_eac[x] else 0)

# Add parvo virus information
df_parvoLab = pd.read_csv(r'../new_data/offshoots/ParvoVirus/DF_parvovirus_LAB.csv', sep=";") # Appointments_REU.csv
print('N patients with parvovirus information: ', len(df_parvoLab['subject_Patient_value'].unique()))
df_parvoLab['valueString'] = df_parvoLab.apply(lambda x : 'Negatief' if x['valueQuantity_value'] == 0 else x['valueString'], axis=1)

d_binarize = {'Negatief' : 0, 'Positief' : 1}

df_parvoLab['ParvoVirus'] = df_parvoLab['valueString'].apply(lambda x : d_binarize[x] if x in d_binarize.keys() else np.nan)
d_parvo = dict(zip(df_parvoLab['subject_Patient_value'], df_parvoLab['ParvoVirus']))
df_treat_new['tested_for_parvo'] = df_treat_new['patnr'].apply(lambda x: 1 if x in d_parvo.keys() else 0)
df_treat_new['ParvoVirus'] = df_treat_new['patnr'].apply(lambda x: d_parvo[x] if x in d_parvo.keys() else np.nan)

# Add HLA subtyping information
df_hla = pd.read_csv(r'../new_data/offshoots/ParvoVirus/DF_HLA_typering_LAB.csv', sep=";") # Appointments_REU.csv
print('N patients with HLA subtyping: ', len(df_hla['subject_Patient_value'].unique()))
df_hla['B27'] = df_hla['valueString'].apply(lambda x : 1 if 'B*27' in x else 0)

# Check hla
d_hla = dict(zip(df_hla['subject_Patient_value'], df_hla['B27']))
d = dict(zip(df_hla[df_hla['B27']==1]['subject_Patient_value'], df_hla[df_hla['B27']==1]['B27']))
for i in d.keys():
    d_hla[i] = d[i]

df_treat_new['tested_for_hla'] = df_treat_new['patnr'].apply(lambda x: 1 if x in df_hla['subject_Patient_value'].unique() else 0)
df_treat_new['B27'] = df_treat_new['patnr'].apply(lambda x: d_hla[x] if x in d_hla.keys() else np.nan)


# Check to see if pseudoIds end up in same cluster
d_longitudinal = {}
for pat in df_treat_new["patnr"].unique():
    d_longitudinal[pat] = len(df_treat_new[df_treat_new['patnr']==pat]['pseudoId'].unique())

df_treat_new['LONGITUDINAL_ID'] = df_treat_new.apply(lambda x: int(x['pseudoId'][-1]) if d_longitudinal[x['patnr']]> 1 else np.nan, axis=1)

# Calculate nr of days between first Mannequin and first consult
df_treat_new['deltaMan'] = pd.to_datetime(df_treat_new['firstMannequin'], format='%Y-%m-%d', errors='ignore').dt.date -pd.to_datetime(df_treat_new['FirstConsult'], format='%Y-%m-%d', errors='ignore').dt.date
df_treat_new['deltaMan'] = df_treat_new['deltaMan'].dt.days

# remove non categorical data from list of categorical
l_cat = [col for col in l_cat if col not in ['FirstConsult', 'patnr', 'pseudoId']]

# Export Metadata
columns = [col for col in df_treat_new.columns if col not in ['pseudoId', 'patnr']] # 'pseudoId',
columns.append('pseudoId')
df_treat_new[columns].to_csv('/exports/reum/tdmaarseveen/RA_Clustering/new_data/5_clustering/df_metadata.csv', index=False)
df_treat_new[['pseudoId', 'patnr']].to_csv('/exports/reum/tdmaarseveen/RA_Clustering/new_data/5_clustering/df_metadata_keys.csv', index=False)
```

# #5.4 Apply PhenoGraph on latent feature space

```python
import pandas as pd
import sys

sys.path.append(r'../../modules/parametric_tsne/')
sys.path.append(r'../../modules/')
sys.path.append(r'../src/1_emr_scripts')

from DeterministicPhenoGraph import phenograph as phenograph2
from DeterministicPhenoGraph.phenograph.cluster import cluster as cluster2
import phenograph
import importlib as imp
import Visualization as func
from umap.parametric_umap import ParametricUMAP
import umap 
from sklearn.manifold import TSNE
import numpy as np
import pickle
import phenograph
imp.reload(func)


# SET PARAMS
neighbours = 270 
SEED = 20221212 #9122022 # 

TECHNIQUE= 'MMAE'



# Latent feature space (a.k.a. Factors from MOFA)
metadata = pd.read_csv('../new_data/5_clustering/df_metadata.csv')
Z = pd.read_csv('../results/embedding/%s.csv' % EMBEDDING_SPACE) # maui_embedding_3.csv
Z = Z[Z.columns[4:]]

# Visualize patient embedding in 2d representation
fit1 = TSNE(random_state=SEED, n_components=2, perplexity=30, metric='minkowski').fit_transform(Z.values)
fit2 = umap.UMAP(random_state=SEED+96812,metric='minkowski', n_neighbors=6).fit(Z.values).embedding_ 

# Perform clustering on top of patient embedding
label, graph, Q  = cluster2(Z.values, k=neighbours, primary_metric='minkowski', n_jobs= 1, seed=SEED) # Do not know why it is seed + 1??

metadata['PhenoGraph_clusters'] = label 
metadata['DAS28(3)'] = func.calculate_DAS28(metadata['TJC'], metadata['SJC'], metadata['BSE'])
metadata['DAS44'] = func.calculate_DAS44(metadata['TJC'], metadata['SJC'], metadata['BSE'])

# Add date of first Consult
metadata['Year_firstConsult'] = metadata['FirstConsult'].str[:4]
metadata['Year_firstConsult'] = metadata['Year_firstConsult'].astype(int) - 2010

# Get more insight into Mannequin by visualizing the ratios
metadata['Swollen ratio'] = metadata['SJC'] / (metadata['SJC']+metadata['TJC'])
metadata['Swollen ratio'] = metadata['Swollen ratio'].fillna(0) 
metadata['Tender ratio'] = metadata['TJC'] / (metadata['SJC']+metadata['TJC'])
metadata['Tender ratio'] = metadata['Tender ratio'].fillna(0)

metadata['Big ratio'] = metadata['Big joints'] / (metadata['TJC']+metadata['SJC'])
metadata['Big ratio'] = metadata['Big ratio'].fillna(0)

metadata['Small ratio'] = metadata['Small joints'] / (metadata['TJC']+metadata['SJC'])
metadata['Small ratio'] =metadata['Small ratio'].fillna(0)

metadata['Symmetrical ratio'] = metadata['Symmetrical joints'] / (metadata['TJC']+metadata['SJC'])
metadata['Symmetrical ratio'] = metadata['Symmetrical ratio'].fillna(0)
metadata = pd.concat([metadata, Z], axis=1) # add latent factors

# Quantify correlation between cluster and physician with Cramers V
print('Correlation between cluster and physician: ', func.cramers_v(metadata['PhenoGraph_clusters'], metadata['Physician']))

func.visualize_umap_bokeh_na_2(fit1, metadata, list(metadata.columns), l_binary=l_cat, patient_id='pseudoId', cluster_id='PhenoGraph_clusters', title='%s_PHENOGRAPH_270_TSNE' % TECHNIQUE)
func.visualize_umap_bokeh_na_2(fit2, metadata, list(metadata.columns), l_binary=l_cat, patient_id='pseudoId', cluster_id='PhenoGraph_clusters', title='%s_PHENOGRAPH_270_UMAP' % TECHNIQUE)


# Save final dataset
metadata.to_csv('../new_data/7_final/%s.csv' % PROJECTION_DATA)
```

#### #5.5.1 Check if clusters are from same time period to ensure similar treatment strategies (addressing bias)

```python
import matplotlib as mpl
import matplotlib.pyplot as plt


def mjrFormatter(x, pos):
    return "$2^{{{0}}}$".format(x)

def mjrFormatter_no_TeX(x, pos):
    return "{0}".format(x+2010)

import seaborn as sns
import matplotlib.pyplot as plt

pal_colorblind = ['#FA4D4D', '#FBC93D',  '#4F6CCF', '#2db9cc', '#4B42FD','#E37E3B', '#C13BE3', '#D55E00' ]

#Colorblind red= '#F46086', 

# Draw a nested boxplot to show bills by day and time
first_ax = sns.violinplot(x="PhenoGraph_clusters", y="Year_firstConsult",
            data=metadata,  palette=pal_colorblind) # hue="smoker", palette=["m", "g"], cmap=mpl.cm.get_cmap('Set1'),
sns.despine(offset=10, trim=True)
first_ax.set_xticklabels(['1','2','3','4'])
first_ax.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(mjrFormatter_no_TeX))
plt.draw()

c1 = metadata[(metadata["PhenoGraph_clusters"]==0) & (~metadata["Year_firstConsult"].isna())]["Year_firstConsult"].values
c2 = metadata[(metadata["PhenoGraph_clusters"]==1) & (~metadata["Year_firstConsult"].isna())]["Year_firstConsult"].values
c3 = metadata[(metadata["PhenoGraph_clusters"]==2) & (~metadata["Year_firstConsult"].isna())]["Year_firstConsult"].values
c4 = metadata[(metadata["PhenoGraph_clusters"]==3) & (~metadata["Year_firstConsult"].isna())]["Year_firstConsult"].values

#stats.kruskal(c1, c2, c3, c4)
```

##### How often tested?

```python
import matplotlib.pyplot as plt

l_parvo_pos = [len(metadata[((metadata['PhenoGraph_clusters']==i)  & (metadata['tested_for_parvo']==1) )]) for i in range(0,4)]
l_parvo_neg = [len(metadata[((metadata['PhenoGraph_clusters']==i)  & (metadata['tested_for_parvo']==0) )]) for i in range(0,4)]

l_parvo_total = list(np.array(l_parvo_neg) + np.array(l_parvo_pos))

df_parvo_results = pd.DataFrame( dict(
    Positive=np.array(l_parvo_pos), # success conventional
    Negative=np.array(l_parvo_neg) )) # total convential
df_parvo_results.index =[1,2,3,4]

# Define Data

df = pd.DataFrame(dict(
    x1=np.array(l_parvo_pos) /np.array(l_parvo_total), # success conventional
    x2=[1,1,1,1] )) # total biological

plt.barh( ['1', '2', '3', '4'], df.x2, align='edge', height=0.4, color='dodgerblue')
plt.barh(['1', '2', '3', '4'], df.x1, align='edge', height=0.4, color='darkblue' )

#plt.yticks(['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'])
plt.ylabel("Clusters")
plt.xlabel("Proportion")
plt.legend([ "Not Tested for Parvovirus",  "Tested for Parvovirus"], bbox_to_anchor=(0.5, 1.05))


for i, (cs, pc) in enumerate(zip(df.iloc[:, 0:].cumsum(1)['x1'],df['x1'])):
    raw_q1, raw_pc1 = list(zip(df_parvo_results.iloc[:, 0:]['Positive'],df_parvo_results['Positive']))[i]
    raw_q2, raw_pc2 = list(zip(df_parvo_results.iloc[:, 0:]['Negative'],df_parvo_results['Negative']))[i]
    if np.round(pc * 100, 1) > 5: # Threshold = 5%
        plt.text(cs - pc / 2, i+ 0.2, str(raw_q1),
            ha = 'center', rotation = 0, fontsize = 8, color='white')
    if np.round((1-pc) * 100, 1) > 5: # Failure
        plt.text(pc + (raw_q2/(raw_q1+raw_q2)) /2, i+ 0.2,  str(raw_q2),
            ha = 'center', rotation = 0, fontsize = 8, color='black')

df_parvo_results
```

# #5.6 Finetuning k from Phenograph [Optional]

Prior to applying phenograph it is important that you selected the optimal k

We will create a heatmap that depicts the similarity between iterations of Phenograph with different k's to select the most optimal k.

```python
import pandas as pd
from scipy import sparse
import numpy as np
import phenograph

# define a function that will give us the relevant output for the input k
def calc_cluster_characteristics(k, embedding, seed):
    results = pd.Series(dtype = 'float64')
    results['k'] = k # metric='minkowski',
    #results['communities'], results['graph'], results['Q'] = phenograph.cluster(Z.values, k=k, seed=SEED+k, clustering_algo='leiden', primary_metric='minkowski')
    #results['communities'], results['graph'], results['Q'] = phenograph.cluster(Z.values, k=k, clustering_algo='leiden',  seed=seed+k) #primary_metric='minkowski') #  
    results['communities'], results['graph'], results['Q'] = cluster2(embedding, k=k, primary_metric='minkowski', seed=seed+k,  n_jobs=1) # metric='minkowski', seed=seed) # random_state=seed,
    results['components'], _ = sparse.csgraph.connected_components(results['graph'], directed = False)
    return results
seed= SEED #20221210#20221209 
ks = np.arange(10, 301, 10) #5 # This is the range of k values we will look at
cluster_chars_list = [calc_cluster_characteristics(k, Z.values, seed) for k in ks]
cluster_characteristics = pd.concat(cluster_chars_list, axis = 1).transpose()
```

```python
from sklearn.metrics import adjusted_rand_score
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
n = len(ks)

Rand_indices = pd.DataFrame(np.zeros((n,n)), index = ks, columns = ks)

Rand_indices.index.name = 'k1'
Rand_indices.columns.name = 'k2'

for i in range(n):
    for j in range(n):
        Rand_indices.iloc[i, j] = adjusted_rand_score(
            cluster_characteristics['communities'][i], 
            cluster_characteristics['communities'][j]
        )
        
cluster_characteristics['clusters'] = cluster_characteristics['communities'].apply(lambda x: len(np.unique(x)))



## Plot heatmap 
fig, ((ax1), (ax2)) = plt.subplots(nrows=2, ncols=1, figsize=(8, 10), sharex='col',
                                                      gridspec_kw={'height_ratios': [16, 4]})
cbar_ax = fig.add_axes([1.05, .388, .02, .4]) # square=True, 
sns.heatmap(Rand_indices, cmap=('coolwarm'), vmin=0, vmax=1, cbar_ax=cbar_ax, xticklabels=False, linewidths=1, ax=ax1).invert_yaxis()

ax2.set_xlabel('k')
ax2.set_ylabel('nr of clusters')

x_ticks = list(Rand_indices.columns)
x_tick_pos = [i + 0.5 for i in range(len(x_ticks))]
ax2.bar(x_tick_pos, cluster_characteristics['clusters'], align='center')
ax2.set_xticks(x_tick_pos)

ax2.set_xticklabels(x_ticks, rotation=90)

plt.tight_layout()
plt.show()
```

## #5.6.3 Robustness over multiple iterations
In less-structured data you would expect a lot of variety in the clustering over multiple iterations. Or data describing patients which lie on a continuum you might find more substantial changes between runs of the PhenoGraph algorithm. Hence, we show the cluster assignments over multiple runs of PhenoGraph, with unchanged parameter settings.

```python
import sys

sys.path.append(r'../../modules/parametric_tsne/')
sys.path.append(r'../../modules/')
sys.path.append(r'../src/1_emr_scripts')
import numpy as np

from DeterministicPhenoGraph import phenograph as phenograph2
from DeterministicPhenoGraph.phenograph.cluster import cluster as cluster2

# leiden algorithm or louvain?
k = 270  # 240 or 210 w/ phenograph lcuster leiden algo
n_runs = 5 # number of times to run PhenoGraph with unchanged parameters

cluster_cols = []
# 0.4516
l_seeds = np.arange(1, n_runs+1, 1) #[1, 2, 3, 4, 5]
for i in range(n_runs): # seed = 2034 # primary_metric='minkowski', 
    communities, graph, Q = cluster2(Z.values, k=k, n_jobs= 1, seed = l_seeds[i], primary_metric='minkowski') # random_state=20220202, 
    cluster_name = 'PhenoGraph_k_' + str(k) + '_run_' + str(i)
    metadata[cluster_name] = pd.Categorical(communities)
    print(len(np.unique(communities)))
    cluster_cols.append(cluster_name) 
```

```python
import numpy as np
from sklearn.metrics import adjusted_rand_score
import seaborn as sns
import matplotlib.pyplot as plt
# 140
Rand_indices_unchanged_params = pd.DataFrame(np.zeros((n_runs,n_runs)))

for i in range(n_runs):
    for j in range(n_runs):
        Rand_indices_unchanged_params.iloc[i, j] = adjusted_rand_score(
            metadata[cluster_cols[i]], 
            metadata[cluster_cols[j]]
        )
        
plt.figure(figsize = (6,6))
sns.heatmap(Rand_indices_unchanged_params, cmap=('coolwarm'), square=True, vmin=0, vmax=1, annot = True).invert_yaxis()
sns.set(font_scale=1)
plt.xlabel('Run A')
plt.ylabel('Run B')
plt.title('Adjusted Rand Score')
plt.show()
plt.close()
```

## #5.8 Sensitivity analysis 
Only keep archetypal patients (removing 10-30% of edge cases), and see if the association with long term outcomes become stronger.

```python
import pandas as pd
from scipy import spatial
import numpy as np

def cycle(my_list, start_at=None):
    start_at = 0 if start_at is None else my_list.index(start_at)
    while True:
        yield my_list[start_at]
        start_at = (start_at + 1) % len(my_list)

# Percentage to prune 
PRUNE = 0.10 

# Save force plot for each archetypal patient
df_arch = pd.DataFrame(metadata[l_coordinates], columns=l_coordinates)
df_arch['PhenoGraph_clusters']  = metadata['PhenoGraph_clusters'].copy()
df_arch['pseudoId'] = metadata['pseudoId'].copy()

l_drop_10 = []
l_drop_20 = []
l_drop_30 = []

d_centroid = {}

cyc = cycle([0,1,2,3])       


for class_idx in range(4):
    indices = df_arch[df_arch['PhenoGraph_clusters']==class_idx].index
    sub_df = df_arch[df_arch['PhenoGraph_clusters']==class_idx][l_coordinates].copy()
    d_centroid[class_idx] = sub_df.median()

for class_idx in range(4):
    indices = df_arch[df_arch['PhenoGraph_clusters']==class_idx].index
    sub_df = df_arch[df_arch['PhenoGraph_clusters']==class_idx][l_coordinates].copy()
    l_sim = []
    for ix, row in sub_df.iterrows():
        
        c1, c2, c3, c4 = next(cyc), next(cyc), next(cyc), next(cyc)
        # Take ratio : distance to own Cluster center  vs smallest distance to anyother cluster
        l_sim.append(spatial.distance.minkowski(d_centroid[c1], row) / min(spatial.distance.minkowski(d_centroid[c2], row),
                    spatial.distance.minkowski(d_centroid[c3], row), spatial.distance.minkowski(d_centroid[c4], row)))
    
    for j in range(round(len(sub_df)*PRUNE)): # remove till you reach top 10
        select_pat = df_arch.iloc[indices[np.argmax(l_sim)]]['pseudoId']
        l_drop_10.append(select_pat)
        l_drop_20.append(select_pat) 
        l_drop_30.append(select_pat) 
        # remove pseudoId from the list
        indices = list(indices)
        indices.remove(indices[np.argmax(l_sim)])
        del l_sim[np.argmax(l_sim)]
        indices = np.array(indices)
        
    for j in range(round(len(sub_df)*0.4)): # remove till you reach top 50%
        select_pat = df_arch.iloc[indices[np.argmax(l_sim)]]['pseudoId']
        l_drop_20.append(select_pat) 
        l_drop_30.append(select_pat) 
        # remove pseudoId from the list
        indices = list(indices)
        indices.remove(indices[np.argmax(l_sim)])
        del l_sim[np.argmax(l_sim)]
        indices = np.array(indices)
        
    for j in range(round(len(sub_df)*0.2)): # remove till you reach top 70%
        select_pat = df_arch.iloc[indices[np.argmax(l_sim)]]['pseudoId']
        l_drop_30.append(select_pat) 
        # remove pseudoId from the list
        indices = list(indices)
        indices.remove(indices[np.argmax(l_sim)])
        del l_sim[np.argmax(l_sim)]
        indices = np.array(indices)


# Add 10% filter
metadata['Thresh_10%'] = metadata['pseudoId'].apply(lambda x : 1 if x in l_drop_10 else 0)
# Add 20% filter
metadata['Thresh_50%'] = metadata['pseudoId'].apply(lambda x : 1 if x in l_drop_20 else 0)
# Add 20% filter
metadata['Thresh_70%'] = metadata['pseudoId'].apply(lambda x : 1 if x in l_drop_30 else 0)

# Save final dataset
metadata.to_csv('../new_data/7_final/MMAE_clustering_270_percCutoff.csv')
```
