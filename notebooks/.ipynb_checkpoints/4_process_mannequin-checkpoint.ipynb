## #4.2 Feature engineering

#### Differentiate between large and small joints
We differentiate between big & small joint based on the EULAR 2010 criteria. 

#### Count nr of symmetrical joints
- Symmetry is asserted based on an exact match for big joint, but fuzzy match for the smaller joints. 
- For example, we can infer symmetry if a patient is affected in MTP-1 on the left and MTP-2 on the right.
- Keywords : schouder, heup, knie, Elleboog/elleboog, enkel (spronggewricht / spronggewicht), mcp, pip, mtp, ip, pols
- We look for a keyword and then we subsequently check if we have inflammation on the left and right


#### Feature engineering
The features are structured according to the STJ format: Symmetrical, Type of inflammation (Tender/Swollen) and Joint (e.g. Knee, MTP, Wrist)

#### Assumptions
Assume : ankle = ( 'onderste spronggewricht', 'bovenste spronggewricht')
- Be careful : if a joint is both tender and swollen it is counted twice!

```python
import numpy as np
# Maybe first sort on date. Then remove duplicates, keep first. Then apply count_special() function 

# Remove artefacts
l_cols_interest = [col for col in final_df.columns if 'OBJECTID' not in col]
final_df = final_df[l_cols_interest].copy()

final_df['Big joints'], final_df['Small joints'], final_df['Symmetrical joints'], final_df['SJC'], final_df['TJC'] = zip(*final_df.apply(lambda x: func.count_special_joints_2(x), axis=1))
final_df = final_df.replace(np.nan, 0)
```

### #4.2.1 Perform One Hot encoding

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# Count data
l_count = ['Big joints', 'Small joints', 'Symmetrical joints', 'SJC', 'TJC']
l_pass = l_count.copy()
l_pass.extend(['FirstConsult', 'patnr','pseudoId'])


# Perform OHE
def cast_to_one_hot(values):
    label_encoder = LabelEncoder()
    integer_encoded = label_encoder.fit_transform(values)

    onehot_encoder = OneHotEncoder(sparse=False)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    df = pd.DataFrame(onehot_encoded, columns=[0, 1])
    df.index = np.arange(0, len(df))
    return df[0], df[1]

# Drop columns without any additional information
final_df = final_df.loc[:, (final_df != final_df.iloc[0]).any()]
l_cat = [col for col in list(final_df.columns) if col not in l_pass]

for cat in l_cat:
    final_df['%s_negative' % (cat)], final_df['%s_positive' % (cat)] = cast_to_one_hot(final_df[cat].values)
    final_df = final_df.drop(columns=cat)

# Drop columns without any additional information
#final_df = final_df.loc[:, (final_df != final_df.iloc[0]).any()]

# write to file
final_df.to_csv('../new_data/4_processed/DF_Mannequin_NEW_Engineered.csv', sep='|', index=False)
final_df
```

### #4.4 Calculate symptom duration

```python
import pandas as pd

df_treat = pd.read_csv(r'../filters/RA_patients_AllNP_Characteristics_counts.csv', sep='|')
df_symptom = pd.read_csv(r'../new_data/4_processed/DF_Symptom_Duration.csv', sep='|', index_col=0)
df_merged = df_symptom.merge(df_treat, how='inner', left_on='patnr_x', right_on='patnr')

# Calculate symptom duration per pseudoId
def getSymptomDuration(row):
    """
    Calculate the symptom duration with respect to the firstConsult.
    Because we use pseudo-Ids it is necessary that we 'pad' the 
    duration with extra days if we look from an extra firstConsult.
    
    Input:
        row = row from the pandas dataframe where the clinical data is merged with 
            the annotated symptom duration file 
    Return
        updated symptom duration according to the date of the first consult
    """
    diff = row['FirstConsult'] - row['time_x']
    diff = diff.days
    return diff + row['amount']

df_merged['time_x'] = pd.to_datetime(df_merged['time_x'], format='%d-%m-%Y', errors='ignore')
df_merged['FirstConsult'] = pd.to_datetime(df_merged['FirstConsult'], format='%Y-%m-%d', errors='ignore')
df_merged['symptomDuration'] = df_merged.apply(lambda x : getSymptomDuration(x), axis=1)

# Add to metadata df
d_symptom = dict(zip(df_merged.pseudoId, df_merged.symptomDuration))

df_treat['symptomDuration'] = df_treat['pseudoId'].apply(lambda x: d_symptom[x] if x in d_symptom.keys() else '' )
df_treat.to_csv(r'../filters/RA_patients_AllNP_Characteristics_symptomDur.csv', sep='|', index=False)
df_treat.head()
```

```python
# import numpy as np
import pandas as pd

# Fill in missing DAS
df_treat_new['DAS28(3)'] = df_treat_new['DAS28(3)'].fillna(df_treat_new.apply(lambda x : func.calculate_DAS28(x['TJC'],x['SJC'], x['BSE']) , axis=1))
df_treat_new['DAS44'] = df_treat_new['DAS44'].fillna(df_treat_new.apply(lambda x : func.calculate_DAS44(x['TJC'],x['SJC'], x['BSE']) , axis=1))

# Add MTP incidence (for quality check)
MTP_JOINTS = [col for col in final_df.columns if 'mtp' in col and 'positive' in col]
df_treat_new['MTP-count'] = final_df[MTP_JOINTS].sum(axis=1)

# Remove duplicate column from df_y
def drop_y(df):
    to_drop = [x for x in df if x.endswith('_y')]
    df.drop(to_drop, axis=1, inplace=True)

drop_y(df_treat_new)

# Keep duplicate column from df_x but remove the suffix
def rename_x(df):
    for col in df:
        if col.endswith('_x'):
            df.rename(columns={col:col.rstrip('_x')}, inplace=True)
            
rename_x(df_treat_new)

def IQR(data):
    # First quartile (Q1)
    if len(data) > 0:
        Q1 = np.percentile(data, 25, interpolation = 'midpoint')

        # Third quartile (Q3)
        Q3 = np.percentile(data, 75, interpolation = 'midpoint')

        # Interquaritle range (IQR)
        iqr = Q3 - Q1
        return [Q1, Q3]
    else :
        return [0, 0]

def summary(lbl, df):
    """
    Calculate summary details of the provided dropped out patients
    
    - DAS
    - FollowUp
    - Symptom duration
    - MTX
    """
    #print('aCCP')
    n = len(df)
    pat = len(df['patnr'].unique())
    #print(df.columns)
    #print(df['RF'].unique())
    rf = '%.0f (%.1f)' % (len(df[df['RF']==1]), (len(df[df['RF']==1])/len(df[df['RF']!=2]))*100)
    rf_neg = '%.0f (%.1f)' % (len(df[df['RF']==0]), (len(df[df['RF']==0]) /len(df[df['RF']!=2]))*100)
    rf_na = '%.0f (%.1f)' % (len(df[df['RF']==2]), (len(df[df['RF']==2])/len(df))*100)
    accp = '%.0f (%.1f)' % (len(df[df['aCCP']==1]), (len(df[df['aCCP']==1])/len(df[df['aCCP']!=2]))*100)
    accp_neg = '%.0f (%.1f)' % (len(df[df['aCCP']==0]), (len(df[df['aCCP']==0]) /len(df[df['aCCP']!=2]))*100)
    accp_na = '%.0f (%.1f)' % (len(df[df['aCCP']==2]), (len(df[df['aCCP']==2])/len(df))*100)
    sex = '%.0f (%.1f)' % (np.sum(df['Sex']), np.mean(df['Sex']) * 100)
    esr = '%.0f (%0.f-%0.f)' % (np.median(df[~df['BSE'].isna()]['BSE']), IQR(df[~df['BSE'].isna()]['BSE'])[0], IQR(df[~df['BSE'].isna()]['BSE'])[1])
    esr_na = '%.1f (%.1f)' % (len(df[df['BSE'].isna()]), (len(df[df['BSE'].isna()])/ len(df))*100)
    age = '%.1f (%.1f)' % (np.mean(df['Age']), np.std(df['Age']))
    sjc = '%.0f (%0.f-%0.f)' % (np.median(df[df['FILTER_MAN_NA_BASELINE']==False]['SJC']), IQR(df[df['FILTER_MAN_NA_BASELINE']==False]['SJC'])[0], IQR(df[df['FILTER_MAN_NA_BASELINE']==False]['SJC'])[1])
    no_sjc = '%.1f (%.1f)' % (len(df[((df['FILTER_MAN_NA_BASELINE']==False) & (df['SJC']==0))]), (len(df[((df['FILTER_MAN_NA_BASELINE']==False) & (df['SJC']==0))])/ len(df[df['FILTER_MAN_NA_BASELINE']==False]))*100)
    sjc_na = '%.0f (%.1f)' % (len(df[df['FILTER_MAN_NA_BASELINE']==True]), (len(df[df['FILTER_MAN_NA_BASELINE']==True])/ len(df))*100) 
    tjc = '%.0f (%0.f-%0.f)' % (np.median(df[df['FILTER_MAN_NA_BASELINE']==False]['TJC']), IQR(df[df['FILTER_MAN_NA_BASELINE']==False]['TJC'])[0], IQR(df[df['FILTER_MAN_NA_BASELINE']==False]['TJC'])[1])
    no_tjc = '%.1f (%.1f)' % (len(df[((df['FILTER_MAN_NA_BASELINE']==False) & (df['TJC']==0))]), (len(df[((df['FILTER_MAN_NA_BASELINE']==False) & (df['TJC']==0))])/ len(df[df['FILTER_MAN_NA_BASELINE']==False]))*100)
    das28 = '%.1f (%.1f-%1.1f)' % (np.median(df[~df['DAS28(3)'].isna()]['DAS28(3)']), IQR(df[~df['DAS28(3)'].isna()]['DAS28(3)'])[0], IQR(df[~df['DAS28(3)'].isna()]['DAS28(3)'])[1])
    das28_na = '%.1f (%.1f)' % (len(df[df['DAS28(3)'].isna()]['DAS28(3)']), (len(df[df['DAS28(3)'].isna()])/ len(df))*100)
    das44 = '%.1f (%.1f-%.1f)' % (np.median(df[~df['DAS44'].isna()]['DAS44']), IQR(df[~df['DAS44'].isna()]['DAS44'])[0], IQR(df[~df['DAS44'].isna()]['DAS44'])[1])
    das44_na = '%.1f (%.1f)' % (len(df[df['DAS44'].isna()]['DAS44']), (len(df[df['DAS44'].isna()])/ len(df))*100)
    proba = '%.3f (%1.f-%1.f)' % (np.median(df['prediction']), IQR(df['prediction'])[0], IQR(df['prediction'])[1])
    #mtp = '%.0f (%0.f-%0.f)' % (np.median(df['MTP-count']), IQR(df['MTP-count'])[0], IQR(df['MTP-count'])[1])
    if 'symptomDuration' in df.columns: 
        symp = '%.0f (%0.f-%0.f)' % (np.median(df[~df['symptomDuration'].isna()]['symptomDuration']), IQR(df[~df['symptomDuration'].isna()]['symptomDuration'])[0], IQR(df[~df['symptomDuration'].isna()]['symptomDuration'])[1])
        symp_na ='%.0f (%0.f)' % (len(df[df['symptomDuration'].isna()]), (len(df[df['symptomDuration'].isna()])/ len(df))*100)
    else :
        symp = np.nan
        symp_na = np.nan
    if 'totalFollowUp' in df.columns:
        fu = '%.f (%0.f-%0.f)' % (np.median(df[~df['totalFollowUp'].isna()]['totalFollowUp']), IQR(df[~df['totalFollowUp'].isna()]['totalFollowUp'])[0], IQR(df[~df['totalFollowUp'].isna()]['totalFollowUp'])[1])
    else :
        fu = np.nan
        # DAS28(3)
    return [lbl, n, pat, sex, age, rf, rf_neg, rf_na, accp, accp_neg, accp_na, esr, esr_na, sjc, no_sjc, sjc_na, tjc, no_tjc, das28, das28_na, das44, das44_na, proba, symp, symp_na, fu]
    

def generatePatientCharacteristics(df_treat):
    """  
    Generate table with baseline characteristics for each step where patients are dropped.
    
    Input:
        df_treat = table with all medication information
    Return:
        df_baseline = generated baseline table
        l_final_set = set for treatment analysis
        l_learning_set = set for learning the representation (this one is larger than final set)
    """
    df_baseline = pd.DataFrame(columns=['id', 'N', 'Pat', 'Sex [n(%)]', 'Age (mean+-SD, yr)', 'RF+ [n(%)]', 'RF- [n(%)]', 'RF_NA [n(%)]', 'aCCP+ [n(%)]', 'aCCP- [n(%)]', 'aCCP_NA [n(%)]', 'ESR (median+-IQR, mm/hr)', 'ESR_NA [n(%)]', 'SJC (median+-IQR)', 'NO_Swollen[n(%)]', 'Swollen_NA[n(%)]', 'TJC (median+-IQR)', 'NO_Tender [n(%)]', 'DAS28(3) (median+-IQR)', 'DAS28(3)_NA [n(%)]', 'DAS44 (median+-IQR)', 'DAS44_NA [n(%)]', 'probability', 'Symptom duration', 'Symptom duration NA [n(%)]', 'Follow-up'])
    print('Unfiltered RA-Patients \t\t| n Pseudo-Ids:', len(df_treat['pseudoId']), '\tn Patnrs:', len(df_treat['patnr'].unique()))
    
    
    # Step 1 : After 2011
    #df = df_treat[df_treat['FILTER_2011']].copy()
    #df_baseline.loc[len(df_baseline)] = summary('Data before 2011', df)
    df_sub = df_treat.copy() # remove layer
    l_learning_set = list(set(df_sub[df_sub['FILTER_LAB_NA_BASELINE']==False]['pseudoId']))
    #print('Data prior to 2011\t\t| n Pseudo-Ids:', len(df_sub['pseudoId']), '\tn Patnrs:', len(df_sub['patnr'].unique()))

    # Step 3 : Medication present at baseline
    df = df_sub[df_sub['FILTER_RX_NA_BASELINE']].copy()
    df_baseline.loc[len(df_baseline)] = summary('RX missing baseline', df)
    df_sub = df_sub[(df_sub['FILTER_RX_NA_BASELINE']==False)].copy() # remove layer
    print('Rx Before Baseline\t\t| n Pseudo-Ids:', len(df_sub['pseudoId']), '\tn Patnrs:', len(df_sub['patnr'].unique()))
    
    # Step 5 : Mannequin present at baseline
    df = df_sub[df_sub['FILTER_MAN_NA_BASELINE']].copy()
    
    #df_baseline.loc[len(df_baseline)] = summary('Mannequin missing baseline', df)
    df_sub = df_sub[(df_sub['FILTER_MAN_NA_BASELINE']==False)].copy() # remove layer
    print('Mannequin Missing at Baseline\t| n Pseudo-Ids:', len(df_sub['pseudoId']), '\tn Patnrs:', len(df_sub['patnr'].unique()))
    
    # Step 7 : Lab present at baseline
    df = df_sub[df_sub['FILTER_LAB_NA_BASELINE']].copy()
    
    df_baseline.loc[len(df_baseline)] = summary('Lab missing at baseline', df) 
    df_sub = df_sub[((df_sub['FILTER_LAB_NA_BASELINE']==False))].copy() # remove layer & (df_sub['potentialRN']==True)
    print('Lab Missing at Baseline\t\t| n Pseudo-Ids:', len(df_sub['pseudoId']), '\tn Patnrs:', len(df_sub['patnr'].unique()))
    
    # Step 8 : final dataset    
    df_baseline.loc[len(df_baseline)] = summary('Final dataset', df_sub)
    
    print('Final set at Baseline\t\t| n Pseudo-Ids:', len(df_sub['pseudoId']), '\tn Patnrs:', len(df_sub['patnr'].unique()))
    l_final_set = list(set(df_sub['pseudoId']))
    
    
    # All dropped patients!
    df = df_treat[~df_treat['pseudoId'].isin(l_final_set)] 
    df_baseline.loc[len(df_baseline)] = summary('Dropped patients', df)
    return df_baseline, l_final_set, l_learning_set

#df_treat_new = pd.read_csv(r'../filters/RA_patients_AllNP_Characteristics.csv', sep='|')
df_baseline, l_final_set, l_learning_set = generatePatientCharacteristics(df_treat_new)
df_baseline.T.to_csv(r'../results/droppedPatients/CharacteristicsDroppedPatients.csv', sep='|')
df_treat_new['FINAL_SELECTION'] = df_treat_new['pseudoId'].apply(lambda x : x in l_final_set)
df_treat_new['LEARNING_SELECTION'] = df_treat_new['pseudoId'].apply(lambda x : x in l_learning_set)

# Also calculate days till treatment - from Mannequin
df_treat_new['MannTillTreat'] = pd.to_datetime(df_treat_new['Lookahead_Treatment'], format='%Y-%m-%d', errors='ignore') - pd.to_datetime(df_treat_new['firstMannequin'], format='%Y-%m-%d', errors='ignore')
df_treat_new['MannTillTreat'] = df_treat_new['MannTillTreat'].dt.days

# Also calculate days till treatment - from PEC (NP)
df_treat_new['DaysTillTreat'] = pd.to_datetime(df_treat_new['Lookahead_Treatment'], format='%Y-%m-%d', errors='ignore') - pd.to_datetime(df_treat_new['FirstConsult'], format='%Y-%m-%d', errors='ignore')
df_treat_new['DaysTillTreat'] = df_treat_new['DaysTillTreat'].dt.days

# Save df_treat_new
df_treat_new.to_csv(r'../filters/RA_patients_AllNP_Characteristics_symptomDur.csv', sep='|', index=False)
df_baseline
```

#### #4.5.4.2 And scaled data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler # Robust z-score
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd

def BinarizeSerology(val, include_na=True):
    """
    Standardize the categorical variables to numerical values
    
    where:
        0 = Negative
        1 = Positive
        2 = Missing
    
    Input:
        val = Categorical value 
        include_na = whether or to include missing values
            - in case the missingness contains information
            - you should keep these values!
    Output:
        val = codified categorical value
    """
    
    if val in ['Positief', 'Sterk pos.', 'Sterk pos', 'Sterk Positief']:
        val = 1
    elif val in ['Dubieus', 'Negatief', 'Zwak pos.', 'Zwak pos', 'Zwak Positief']:
        val= 0
    elif include_na:
        if val in ['Stopgezet', '-volgt-', '@volgt', 'Niet doorgeg', 'Niet het jui', 
                   'gestopt', 'Bepaling niet ui', ' ', 'Geen analyse mog', 'Te weinig materi', 
                   'Geen materiaal o', 'te weinig ma', 'gestopt',  'Geen analyse', 'geen uitslag', np.nan]:
            val = np.nan
    return val

def normalizeLab(df):
    """
    Input: 
        df = complete dataframe
    
    Description:
        This function processes the Lab data. 
        Categorical data is subjected to a one-hot-encoding
        while the numerical data is cast to a z-scale as normalization
        
        Furthermore, the ESR (or BSE in dutch) is subjected to a 
        log-transformation.
    
    Output:
        df_dummies = dataframe with preprocessed categorical data 
        df_b = dataframe with preprocessed numerical data
    """
    l_num = ['Leuko', 'MCH', 'Hb', 'Ht', 'MCHC', 'MCV', 'Trom',  'Age', 'BSE'] # 'BSE', 'Lym', 'Mono', 
    l_cat = ['RF', 'aCCP', 'Sex']
    
    # Log transform the ESR data
    df['BSE'] = np.log(df['BSE'])
    
    #df[l_num] = df[l_num].apply(lambda x : np.log(x)) # 
    
    ct = ColumnTransformer([("num", PowerTransformer(), l_num)], remainder="passthrough") #
    df_num= pd.DataFrame(ct.fit_transform(df[l_num]), columns=l_num)
    
    df_cat = df[l_cat].copy()
    
    def cast_to_one_hot(values):
        # integer encode

        label_encoder = LabelEncoder()
        integer_encoded = label_encoder.fit_transform(values)
        
        onehot_encoder = OneHotEncoder(sparse=False)
        integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
        onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
        if len(np.unique(values)) == 3:
            df = pd.DataFrame(onehot_encoded, columns=[0, 1,2])
            df.index = np.arange(0, len(df))  
            return df[0], df[1], df[2] 
        elif len(np.unique(values)) == 2:
            df = pd.DataFrame(onehot_encoded, columns=[0, 1])
            df.index = np.arange(0, len(df))  
            return df[0], df[1], None  
        elif len(np.unique(values)) < 2:
            return None, None, None  

    def cast_to_str(x):
        d = {1: 'positive', 0:'negative', 2: 'missing'}
        return d[x]

    CATEGORICAL_PROCESSING = False
    # Categorical preprocessing
    if CATEGORICAL_PROCESSING:
    
        for cat in l_cat:
            print(cat)
            print(df_cat[cat].unique())
            # This will give an error if your index doesnt count from 0 to n
            df_cat['%s_negative' % (cat)], df_cat['%s_positive' % (cat)], df_cat['%s_missing' % (cat)] = cast_to_one_hot(df_cat[cat].values)
            #print(matrix[cat + '_negative' ])
            if len(df_cat[df_cat[cat + '_negative' ].isna()])>1: # remove if column doesnt hold any information:
                df_cat = df_cat.drop(columns=['%s_negative' % (cat), '%s_positive' % (cat), '%s_missing' % (cat)])
                print("Column %s doesn't hold any information and is therefore removed" % (cat))
            df_cat = df_cat.drop(columns=cat)
        df_cat = df_cat.dropna(axis=1, how='all')
    df_imp = df_cat.merge(df_num, left_index=True, right_index=True)
    
    df_imp = df_imp.merge(df[['pseudoId', 'time']], left_index=True, right_index=True)
    return df_imp
    

# Encode antigen test labels
l_cat = ['RF', 'aCCP', 'Sex'] 
for cat in l_cat:
    df_lab[cat] = df_lab[cat].apply(lambda x : BinarizeSerology(x, include_na=True))

# Normalize lab values
df_lab = normalizeLab(df_lab)

l_lab = [col for col in df_lab.columns ] # ['pseudoId', 

df_lab[l_lab].to_csv('../new_data/5_clustering/df_lab_scaled_demographics.csv', index=False, sep=',')
df_lab[l_lab].head()
```
