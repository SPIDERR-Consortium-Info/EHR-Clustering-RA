## #6.1 Preprocessing Medicator data


```python
import re
import numpy as np
import pandas as pd

# import Medicator
df_med = pd.read_csv('/exports/reum/tdmaarseveen/RA_Clustering/new_data/1_raw/Medication_REU.csv', sep=';', header=None)
df_med.columns = ['PATNR', 'periodOfUse_valuePeriod_start_date', 'periodOfUse_valuePeriod_end_date', 'usageDuration_valueDuration_value', 'dosageInstruction_text', 'dosageInstruction_additionalInstruction_text', 'dosageInstruction_timing_repeat_frequency', 'dosageInstruction_timing_repeat_period', 'dosageInstruction_route_display', 'dosageInstruction_doseQuantity_value', 'dosageInstruction_doseQuantity_unit_display_original', 'code_text', 'ATC_code', 'ATC_display', 'ATC_display_nl']

def extract_dose(row):
    """ 
    Extract the dose embedded in the code_text column. 
    We multiply by quantity in case a patient recieves multiple
    'portions' in a day
    
    Input:
        row = single row from medicator
    Output:
        dose = dose retrieved from the description
        unit = quantity unit 
        one_off = whether or not the drug is prescribed as a single use.
    """
    
    dose = np.nan
    unit = np.nan
    one_off = False
    ambiguous = False
    quantity = str(row['quantity']).replace(',', '.')
    
    # Check if only a single use 
    if 'eenmalig' in row['instruction']:
        one_off = True

    code_instruction = row['instruction']
    l_cand  = re.findall('(\(?)(\d+\,?\.?\d+)(MG|mg|milligram|ml)', code_instruction)
    if len(l_cand) == 0:
        code_text = row['codeText']
        l_cand = re.findall('(\(?)(\d+\,?\d+)(MG)', code_text)
    
    for cand in l_cand:
        if cand[0] != '(':
            dose, unit = float(cand[1].replace(',', '.')), cand[2]
    
    # Calculate total concentration if you have multiple pills
    if row['unit'] == 'STUK':
        if quantity != '0' and quantity != np.nan: 
            dose *= float(quantity)
        
    # Just use values directly from columns  
    elif row['route']!='SUBCUTAAN' and row['route']!='PARENTERAAL' and \
    row['drugName']!='HYDROXYCHLOROQUINE' and one_off == False and row['quantity']!='0':
        dose = row['quantity']
        unit = row['unit']
    d_trans = {'MG' : 'MILLIGRAM', 'ml' : 'MILLILITER', 'mg' : 'MILLIGRAM', 'MILLIGRAM' : 'MILLIGRAM'}
    if unit in d_trans:
        unit = d_trans[unit]#'MILLIGRAM'
    return dose, unit, one_off

def format_medication_table(df):
    cols = ['PATNR', 'periodOfUse_valuePeriod_start_date', 'periodOfUse_valuePeriod_end_date', 'usageDuration_valueDuration_value', 'dosageInstruction_text', 'dosageInstruction_timing_repeat_frequency', 'dosageInstruction_timing_repeat_period', 'dosageInstruction_route_display', 'dosageInstruction_doseQuantity_value', 'dosageInstruction_doseQuantity_unit_display_original', 'code_text', 'ATC_display_nl']
    new_cols = ['patnr', 'startDate', 'endDate', 'duration', 'instruction', 'frequencyUnit', 'repeatPeriod',  'route', 'quantity', 'unit', 'codeText', 'drugName']

    df = df[cols].copy()
    df.columns = new_cols
    
    df['extractDose'], df['extractUnit'], df['oneOff'] = zip(*df.apply(lambda x: extract_dose(x), axis=1))
    return df 

df_med = df_med.fillna('')
df_med = format_medication_table(df_med)

df_med.to_csv('/exports/reum/tdmaarseveen/RA_Clustering/new_data/6_medication/Clustering_Medicator_formatted.csv', sep=';', index=False)
df_med
```

### #6.2.1 Combine Treatment information with Disease activity (DAS)

WARNING : THIS MAY TAKE A WHILE  

- These two tables are linked to examine the treatment response
- In our analysis we look from the date that patients receive their first prescription onwards (this is important for the interpretation of the subsequent survival analysis) 

```python
l_final_pat = list(df_treat_new['patnr'].unique())
df_med = df_med[df_med['patnr'].isin(l_final_pat)]
df_das = df_das[df_das['patnr'].isin(l_final_pat)]

len(df_das['patnr'].unique())
```

```python
import pandas as pd
import numpy as np

def create_treatment_composite(df_med, df_das, d_pseudo_date, d_pseudo_pat):
    """
    Create a table for downstream analysis of treatment response
    
    Input:
        - df_med = processed table with Medication information
        - df_das = processed table with DAS information
        - d_pseudo_date = dictionary of pseudo ids linked to first dates
        - d_pseudo_pat = dictionary of pseudo-ids linked to patient ids
    Output:
        - df_treat = composite table with DAS28(3) combined with current treatment information
    """
    
    df_treat = pd.DataFrame(columns=['pseudoId', 'patnr', 'treatID', 'Drug', 'Dosage', 'Unit', 'Instruction', 'StartDateTreatment', 'EndDateTreatment', 'daysTreatment', 'DAS28(3)', 'dateDAS', 'daysDAS', 'SJC', 'TJC', 'ESR', 'extractDate']) # columns=['patnr', 'DAS28(3)', 'CurTreatment', 'Dosage', 'date', 'days', 'daysCurTreatment'
    tol = 1
    
    for pid in d_pseudo_pat.keys():
        print(pid)
        no_med = True # register if medication is found
        missing_baseline = False
        patnr = d_pseudo_pat[pid]
        
        ix = 0
        
        # Create subset of medication table
        sub_df = df_med[df_med['patnr']==patnr].copy()
        sub_df = sub_df.sort_values(by='startDate')
        
        # Create subset of disease activity table
        sub_das = df_das[df_das['patnr']==patnr].copy()
        
        # Get first date associated with the pseudoId
        first_date = pd.to_datetime(d_pseudo_date[pid], format='%Y-%m-%d', errors='ignore')
        
        l_das = list(sub_das['days'].values)
        l_med = []
        
        for ix, drug in enumerate(sub_df['drugName'].values):
            startDate = sub_df['startDate'].iloc[ix]
            endDate = sub_df['endDate'].iloc[ix]
            cor_das = sub_das[(pd.to_datetime(startDate, format='%Y-%m-%d', errors='ignore') - pd.DateOffset(days=tol) <= pd.to_datetime(sub_das.date, format='%Y-%m-%d', errors='ignore')) & (pd.to_datetime(endDate, format='%Y-%m-%d', errors='ignore') + pd.DateOffset(days=tol) >= pd.to_datetime(sub_das.date, format='%Y-%m-%d', errors='ignore'))]#.iloc[0]
            delta_days = pd.to_datetime(startDate, format='%Y-%m-%d', errors='ignore') - first_date
            delta_days = delta_days.days
            for index, match in cor_das.iterrows():
                df_treat.loc[len(df_treat)] = [pid, patnr, '%s_%s' % (patnr, ix), sub_df['drugName'].iloc[ix], sub_df['extractDose'].iloc[ix], sub_df['extractUnit'].iloc[ix], sub_df['instruction'].iloc[ix], sub_df['startDate'].iloc[ix], sub_df['endDate'].iloc[ix], delta_days, match['DAS28(3)'], match['date'], match['days'], match['SJC'], match['TJC'], match['ESR'], sub_df['startDate'].iloc[ix]]
                if match['days'] in l_das: # Keeping track of patients with & without a DAS
                    l_das.remove(match['days'])
            if len(cor_das) == 0: # bookkeeping Medication
                l_med.append(ix)
        for days in l_das:
            match = sub_das[sub_das['days']==days].iloc[0]
            df_treat.loc[len(df_treat)] = [pid, patnr, np.nan, 'NO MEDICATION', np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, match['DAS28(3)'], match['date'], match['days'], match['SJC'], match['TJC'], match['ESR'], match['date']]
        for ix in l_med:
            df_treat.loc[len(df_treat)] = [pid, patnr, '%s_%s' % (pid, ix), sub_df['drugName'].iloc[ix], sub_df['extractDose'].iloc[ix], sub_df['extractUnit'].iloc[ix], sub_df['instruction'].iloc[ix], sub_df['startDate'].iloc[ix], sub_df['endDate'].iloc[ix], delta_days, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, sub_df['startDate'].iloc[ix]]
    
    df_treat = df_treat.sort_values(by='extractDate').reset_index(drop=True)
    return df_treat 

# Create composite tables
df_treat = create_treatment_composite(df_med, df_das, d_pseudo_date, d_pseudo_pat)
df_treat.to_csv('/exports/reum/tdmaarseveen/RA_Clustering/new_data/7_final/TreatmentResponseTable.csv', sep=';', index=False)
```

```python
import re

def IQR(data):
    # First quartile (Q1)
    Q1 = np.percentile(data, 25, interpolation = 'midpoint')

    # Third quartile (Q3)
    Q3 = np.percentile(data, 75, interpolation = 'midpoint')
   
    # Interquaritle range (IQR)
    iqr = Q3 - Q1
    return [Q1, Q3]

def create_baselineTable(df_cluster, df_das, df_treat, col_cluster, FOLLOWUP=365):
    """
    Create a baseline table and display all important cluster characteristics
    as well as the outcome characteristics after 1 year (e.g. DAS28 / DAS44)
    
    To infer the DAS response, we take the DAS after 1 year with 
    a +/- 3 months tolerance window.
    
    Input:
        df_cluster = table with lab/mannequin baseline variables and final clusters 
        df_das = table with DAS scores
        df_treat = table with treatment
        col_cluster = the column in df_cluster that features the cluster ids
        FOLLOWUP = define time window for response
        # treatment switch?
    Output:
        df_baseline = 
        
    ToDo:
        N = number of DAS != number of medication
    """
    c_med_present = 0
    l_miss_tot = 0
    l_miss_final = 0
    l_miss_symp = 0
    df_baseline = pd.DataFrame(columns=['cluster', 'N', 'Sex [n(%)]', 'RF [n(%)]', 'aCCP [n(%)]', 'DAS44 (median+-IQR)', 'SJC (median+-IQR)', 'TJC (median+-IQR)', 'ESR (median+-IQR, mm/hr)', 'Age (mean+-SD, yr)', 'DAS28 (median+-IQR)', 'final_DAS28 (median+-IQR)', 'prevalence_TreatSwitch [n(%)]', 'MTX [n(%)]', 'Follow up (median+-IQR, days)', 'Small joints (median+-IQR)', 'Big joints (median+-IQR)', 'Symmetrical joints (median+-IQR)', 'Symptom duration (median+-IQR, days)', 'MCV (median+-IQR)', 'MCHC (median+-IQR)'])
    d_response = {} # patnr; [begin, end]
    d_firstSwitch = {}
    d_totalFollowUp = {}
    
    def calculate_DAS28(tjc, sjc, esr):
        """
        Calculate DAS28 with 3 variables : TJC, SJC and ESR (BSE)
        """
        if esr != 0:
            das28 = (0.56 * np.sqrt(tjc) + 0.28 * np.sqrt(sjc) + 0.70 * np.log(esr)) * 1.08 + 0.16
        else :
            print(esr, tjc, sjc)
            print(eql)
        return das28

    def calculate_DAS44(tjc, sjc, esr):
        """
        Calculate DAS44 with 3 variables : RAI, SJC and ESR (BSE)
        """

        das44= (0.53938 * np.sqrt(tjc) + 0.0650 * (sjc) + 0.330 * np.log(esr)) + 0.224 #0.224
        return das44
    
    for cluster in df_cluster[col_cluster].unique():
        sub_df = df_cluster[df_cluster[col_cluster]==cluster].copy() # seurat output
        
        # Features to extract for each cluster
        l_das = []
        l_sjc = []
        l_tjc = []
        l_rf = []
        l_accp = []
        l_esr = []
        l_age = []
        l_sex = []
        l_das28 = []
        l_das28_final = [] 
        l_sjc_final = [] 
        l_tjc_final = [] 
        l_response = [] # ' Unknown', 'Positive', 'Negative'?
        l_mtx = []
        l_shift = [] # 0=no, 1=yes
        l_followup = []
        l_first_switch = []
        l_small_joints = []
        l_big_joints = []
        l_mcv = []
        l_mchc = []
        l_symmetrical_joints = []
        l_symptom_duration = []
        c_response = 0
        
        # Loop through each patient within a cluster
        for pid in sub_df[sub_df['pseudoId'].isin(df_das['pseudoId'].unique())]['pseudoId'].unique(): 
            #print(pid)
            # Either take days or new_days..
            # Beware : baseline characteristics are collected with the DAS table (a.k.a. mannequin as a reference)
            df_pid = df_das[((df_das['pseudoId']==pid))].copy()
            
            # Create a sub table with Medication information for a single patient
            df_submed = df_treat[((df_treat['pseudoId']==pid))].copy()
            df_submed = df_submed.sort_values(by='StartDateTreatment')
            
            # Beware function cant interpret drugnames with special characters
            df_submed['Drug'] = df_submed['Drug'].apply(lambda x : str(x).replace(')', '_').replace('(', '_').replace(' ', '_')) #.str.replace(') \(', '_')
            
            d_med = df_submed.groupby('StartDateTreatment')['Drug'].apply(list).to_dict()
            
            
            
            # calculate follow up according to DAS table
            d_totalFollowUp[pid] = (pd.to_datetime(df_pid['date'].max(), format='%Y-%m-%d', errors='ignore') - pd.to_datetime(df_pid['date'].min(), format='%Y-%m-%d', errors='ignore')).days

            # Tolerance of 3 months
            tol = 93
            
            med_present = False
            
            
            def get_first_row(df_pid):
                # Take the most recent patient characteristics before the start of treatment
                if 0 in df_pid['days_in_treat'].unique():
                    # date treatment started
                    return df_pid[df_pid['days_in_treat']==0].iloc[0]
                elif any(t < 0 for t in list(df_pid['days_in_treat'].unique())):
                    # closest to the date treatment started
                    l_neg = [t for t in list(df_pid['days_in_treat'].unique()) if t < 0]
                    return df_pid[df_pid['days_in_treat']==max(l_neg)].iloc[0]
                else :
                    # alternatively, take the most early DAS
                    return df_pid.sort_values(by='days_in_treat').iloc[0]
            
            row = get_first_row(df_pid)
            #print(row, )
            
            if len(d_med) > 0:
                # Check if smaller than PEC + tolerance (tol2) and bigger than TREATPEC
                # Or bigger than PEC and smaller than TREATPEC
                tol2 = 93
                
                # Get starting medication
                l_first_date = [i for i in d_med.keys() if (pd.to_datetime(i, format='%Y-%m-%d', errors='ignore') <= pd.to_datetime(row['TREATPEC'], format='%Y-%m-%d', errors='ignore').tz_localize(None) + pd.DateOffset(days=tol2)) and (pd.to_datetime(i, format='%Y-%m-%d', errors='ignore') >= pd.to_datetime(row['TREATPEC'], format='%Y-%m-%d', errors='ignore').tz_localize(None))]
                
                med = [medi for i in l_first_date for medi in d_med[i]]#[-1]
                med = list(np.unique(med)) 

                if 'METHOTREXAAT' in med:
                    l_mtx.append(1)
                else :
                    l_mtx.append(0)
                    
                def getSwitchDate(l_new_med, d_med):
                    """
                    l_new_med = list with new medication (intersection)
                    d_med = dictionary with all medication
                    """
                    
                    new_med = [k for k,v in d_med.items() if type(re.search("(" + "|".join(l_new_med) + ")", str(v))) != type(None)] # sloww
                    l_dates = new_med#.keys()
                    switch_date = min(l_dates)
                    return switch_date
                    
                # Get index of first switch!
                ##########################################################################
                delta = np.nan
                for index, rec in df_submed.iterrows(): # check all prescriptions of a specific patient
                    last_med = []
                    tol3 = 62 # chunk entries within 2 months from each other and search within a chunk to scan for newly introduced med
                    event = False
                    # d_med
                    end_date_med = rec['StartDateTreatment'] 
                    l_last_date = [i for i in d_med.keys() if (pd.to_datetime(i, format='%Y-%m-%d', errors='ignore') < pd.to_datetime(end_date_med, format='%Y-%m-%d', errors='ignore') + pd.DateOffset(days=tol3)) and (pd.to_datetime(i, format='%Y-%m-%d', errors='ignore') > pd.to_datetime(end_date_med, format='%Y-%m-%d', errors='ignore') - pd.DateOffset(days=tol3))]
                    pot_last_med = [medi for i in l_last_date for medi in d_med[i]]
                    if list(np.unique(pot_last_med)) != []:
                        last_med = list(np.unique(pot_last_med))

                    #print('DATES:', rec['StartDateTreatment'], df_pid['PEC'].iloc[0])
                    if [i for i in last_med if i not in ['PREDNISOLON', 'METHYLPREDNISOLON']]!=[i for i in med if i not in ['PREDNISOLON', 'METHYLPREDNISOLON']] and pd.to_datetime(rec['StartDateTreatment'], format='%Y-%m-%d', errors='ignore') >= pd.to_datetime(df_pid['PEC'].iloc[0], format='%Y-%m-%d', errors='ignore') :#  and rec['daysTreatment'] != 0  #  and last_med != []  and last_med != []
                        #print(pid)
                        if [i for i in list(set(last_med) - set(med)) if i not in ['PREDNISOLON', 'METHYLPREDNISOLON']] == []:
                            #print('FILTERED BECAUSE THERE IS NO REAL TREATMENT SWITCH HERE (just MTX/PRED)')
                            # There is no real treatment switch -> patient starts mtx / prednison
                            break
                        l_new_med =  list(set(last_med) - set(med)) # .symmetric_difference( 
                        l_new_med = [i for i in l_new_med if i not in ['PREDNISOLON', 'METHYLPREDNISOLON']]
                        
                        # Get actual first date
                        if pid == '304852167_0':
                            print('Start_treatment:', rec['StartDateTreatment'], 'PEC:', df_pid['PEC'].iloc[0] )
                            print('Previous:', med, 'New:', last_med)
                            print(l_new_med)
                            print(getSwitchDate(l_new_med, d_med))
                        delta =pd.to_datetime(getSwitchDate(l_new_med, d_med), format='%Y-%m-%d', errors='ignore') - pd.to_datetime(df_pid['TREATPEC'].iloc[0], format='%Y-%m-%d', errors='ignore') # .tz_localize('UTC')
                        delta = delta.days
              
                        # rec['daysTreatment']
                        treatment_shift = True
                        break 
                d_firstSwitch[pid] = delta
                l_first_switch.append(delta)
                ##########################################################################
                
                
                if last_med != med and med != np.nan:
                    # ignore a change in prednisolon or methylprednisolon
                    if [i for i in last_med if i not in ['PREDNISOLON', 'METHYLPREDNISOLON']]==[i for i in med if i not in ['PREDNISOLON', 'METHYLPREDNISOLON']]: 
                        treatment_shift= False
                        l_shift.append(0)
                    else :
                        treatment_shift =True
                        l_shift.append(1)
                else :
                    treatment_shift= False
                    l_shift.append(0)
                med_present = True
                
            else :
                print('No treatment information for pid %s' % str(pid))
                med = np.nan
                d_firstSwitch[pid] = np.nan
            # Register patient details if we have medication information available within the year:
            if med_present == True:
                if len(d_med) > 0:
                    d_response[pid] = [med, last_med, treatment_shift, ('METHOTREXAAT' in med)] # days
                else :
                    d_response[pid] = [[], [], False, False]
            else :  ### COUNT NR OF PATIENTS WITH MEDICATION
                c_med_present += 1
            
            
            # Get treatment response after 1 year + tolerance
            # 3 months before
            df_response = df_pid[(df_pid['new_days']>FOLLOWUP-tol)].copy()
            
            # 3 months after
            df_response = df_response[(df_response['new_days']<FOLLOWUP+tol)]
            
            # Get row with treatment response (~1 year +/- 3 mo after baseline)
            # Response is switch so you don't want to limit yourself 1 year
            # rather determine first date and look from there
            if len(df_response) > 0:
                # get row closest to 1 year
                response_row = df_response.loc[abs(df_response['new_days'] - FOLLOWUP).idxmin()]  # days
                
            else : 
                # Get treatment response (tolerance 3 months before)
                df_response = df_pid[(df_pid['new_days']>FOLLOWUP-tol)].copy()
                
                # Get treatment response (tolerance: 3 months after)
                df_response = df_response[(df_response['new_days']<FOLLOWUP+tol)]
                if len(df_response) > 0:
                    response_row = df_response.loc[abs(df_response['new_days'] - FOLLOWUP).idxmin()] 
                
            #print('Length:', len(df_response))
            
            # Only register data from subjects that have medication data within a year
            if med_present == True:
                row_2 = sub_df[sub_df['pseudoId']==pid].iloc[0] # Get baseline data (same as SEURAT input data)
                l_sjc.append(row_2['SJC'])
                l_tjc.append(row_2['TJC'])
                if row_2['aCCP'] != 2: # missing
                    l_accp.append(row_2['aCCP'])
                if row_2['RF'] != 2: # missing
                    l_rf.append(row_2['RF'])
                    
                l_small_joints.append(row_2['Small joints'])
                l_big_joints.append(row_2['Big joints'])
                if row_2['symptomDuration'] == row_2['symptomDuration']:
                    l_symptom_duration.append(row_2['symptomDuration'])
                else :
                    l_miss_symp += 1
                l_symmetrical_joints.append(row_2['Symmetrical joints'])
                l_age.append(row_2['Age'])
                l_sex.append(row_2['Sex'])
                l_esr.append(row_2['BSE'])
                l_mcv.append(row_2['MCV'])
                l_mchc.append(row_2['MCHC'])
                
                # Calculate DAS at baseline
                l_das28.append(calculate_DAS28(row_2['TJC'], row_2['SJC'], row_2['BSE']))
                l_das.append(calculate_DAS44(row_2['TJC'], row_2['SJC'], row_2['BSE']))
                d_response[pid].append(calculate_DAS44(row_2['TJC'], row_2['SJC'], row_2['BSE']))
                d_response[pid].append(calculate_DAS28(row_2['TJC'], row_2['SJC'], row_2['BSE']))
                if len(df_response) > 0:
                    d_response[pid].append(response_row['DAS28(3)'])
                    c_response += 1
                else :
                    if row['days_in_treat']<=7:
                        l_miss_final += 1
                    d_response[pid].append(np.nan)

                if len(df_response) > 0: # len(d_med) > 0: #if len(d_med) > 0: -> error response row
                    d_response[pid].append(response_row['new_days'])
                else :
                    d_response[pid].append(np.nan)

                if len(df_response) > 0:
                    if d_response[pid][6] >= 0  and len(d_response)>1: 
                        l_das28_final.append(d_response[pid][6])
                if d_response[pid][7] is not np.nan:
                    l_followup.append(d_totalFollowUp[pid])

                d_response[pid].append(cluster) 
            else :
                row_2 = sub_df[sub_df['pseudoId']==pid].iloc[0] 
                print('Patient doesnt have medication:', row_2['pseudoId'], ' (Run #6.2.1 for new extraction)')
                

        l_miss_tot += len(sub_df['pseudoId'].unique()) - len(l_das)
        print('Cluster %s (Total n= %s; With initial DAS = %s; With complete DAS= %s)' % (cluster, len(sub_df[sub_df['pseudoId'].isin(df_das['pseudoId'].unique())]), len(l_das), str(c_response)))
        df_baseline.loc[len(df_baseline)] = [cluster, '%.2f' % len(l_shift), '%.0f (%.1f)' % (np.sum(l_sex), np.mean(l_sex) * 100), '%.0f (%.1f)' % (np.sum(l_rf), np.mean(l_rf) * 100), '%.0f (%.1f)' % (np.sum(l_accp), np.mean(l_accp) * 100), '%.1f (%.1f-%.1f)' % (np.median(l_das), IQR(l_das)[0], IQR(l_das)[1]) , '%.0f (%.0f-%.0f)' % (np.median(l_sjc), IQR(l_sjc)[0], IQR(l_sjc)[1]), '%.0f (%.0f-%.0f)' % (np.median(l_tjc), IQR(l_tjc)[0], IQR(l_tjc)[1]), '%.0f (%.0f-%.0f)' % (np.median(l_esr), IQR(l_esr)[0], IQR(l_esr)[1]), '%.1f (%.1f)' % (np.mean(l_age), np.std(l_age)), '%.1f (%.1f-%.1f)' % (np.median(l_das28), IQR(l_das28)[0], IQR(l_das28)[1]), '%.1f (%.1f-%.1f)' % (np.median(l_das28_final), IQR(l_das28_final)[0], IQR(l_das28_final)[1]), '%.0f (%.1f)' % (np.sum(l_shift), np.mean(l_shift) * 100), '%.0f (%.1f)' % (np.sum(l_mtx), np.mean(l_mtx) * 100), '%.0f (%.0f-%.0f)' %  (np.median(l_followup), IQR(l_followup)[0], IQR(l_followup)[1]), '%.0f (%.0f-%.0f)' % (np.median(l_small_joints), IQR(l_small_joints)[0], IQR(l_small_joints)[1]), '%.0f (%.0f-%.0f)' % (np.median(l_big_joints), IQR(l_big_joints)[0], IQR(l_big_joints)[1]), '%.0f (%.0f-%.0f)' % (np.median(l_symmetrical_joints), IQR(l_symmetrical_joints)[0], IQR(l_symmetrical_joints)[1]), '%.0f (%.0f-%.0f)' % (np.median(l_symptom_duration), IQR(l_symptom_duration)[0], IQR(l_symptom_duration)[1]), '%.1f (%.1f-%.1f)' % (np.median(l_mcv), IQR(l_mcv)[0], IQR(l_mcv)[1]), '%.1f (%.1f-%.1f)' % (np.median(l_mchc), IQR(l_mchc)[0], IQR(l_mchc)[1])]
        
    df_baseline = df_baseline.set_index('cluster')
    print('Missing Medication for %s entries' % str(c_med_present))
    print('Missing initial DAS for %s entries' % str(l_miss_tot))
    print('Missing Symptom duration for %s entries' % str(l_miss_symp))
    print('Missing final DAS for %s entries:' % str(l_miss_final))
    
    return df_baseline.T, d_response, d_firstSwitch, d_totalFollowUp

# Check for MTX-starter?
df_MTX = pd.read_csv('/exports/reum/tdmaarseveen/RA_Clustering/new_data/7_final/SummaryResponseTable_MMAE.csv', sep=';')[['patnr', 'MTX-starter']] # 5
d_MTX = dict(zip(df_MTX['patnr'], df_MTX['MTX-starter']))
df_cluster['MTX-starter'] = df_cluster['pseudoId'].apply(lambda x : d_MTX[x] if x in d_MTX else False)

# Check for stringent population (PPV>.95)
# df_cluster[df_cluster['probability']>.99]
df_baseline, d_response, d_firstSwitch, d_totalFollowUp =create_baselineTable(df_cluster[df_cluster['MTX-starter']==True], df_das, df_treat, col_cluster='PhenoGraph_clusters') #

# Export table to csv
df_baseline = df_baseline[[0, 1, 2, 3]] # sort columns , 4
df_baseline.to_csv('/exports/reum/tdmaarseveen/RA_Clustering/new_data/7_final/BaselineTable_MMAE_MTX_ONLY.csv', sep=';')
```
